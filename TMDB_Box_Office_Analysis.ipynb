```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 

#print(os.getcwd())

df = pd.read_csv("/content/drive/MyDrive/Data Science/Projects/TMDB/train.csv")
display(df.head(5))

display(df.info())
display(df.describe())
```

```python
df['original_language'].value_counts()
```

# Data Visualization Begins

```python
print(type(df.head(10)))

df_first10 = df.head(10)

display(df_first10)
```

##How many of these movies are in English?

```python
plt.figure(figsize=(24,10))
sns.histplot(x="original_language", data=df, color="mediumblue")
plt.yticks(np.arange(0, 2700, 100))

plt.grid()
plt.show()
```

```python
df['original_language'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(20,11))
sns.scatterplot(x='budget', y='revenue', hue='original_language', data=df)

plt.show()
```

## Are English movies more expensive?

```python
plt.figure(figsize=(20,8))
plt.bar(x="original_language", height="budget", data=df)

plt.xlabel("Languages", fontsize=14)
plt.ylabel("Budget in 10^8 Units", fontsize=14)

plt.grid()
plt.tight_layout()
plt.show()
```

## What is the Budget Range of these movies?

```python
df_first10 = df.head(1000)

plt.figure(figsize=(20,10))
sns.histplot(x="budget", data=df_first10)
plt.ylabel("Number of Movies")
plt.yticks(np.arange(0, 451, 50))
#plt.xticks(np.arange(0,25000000, 100000))

plt.grid()
plt.tight_layout()
plt.show()
```
